```python
import pandas
import numpy
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
```

###  for 2**20 size array

```python
no=[1,2,4,8,16]
t=[0.0486912,0.026366,0.0162837,0.0128029,0.0113025]
```

```python
plt.plot(no,t)
plt.title('Time for a fixed array of 2**20');
plt.xlabel('Threads')
plt.ylabel('Time in seconds')
plt.grid()
```

```python
ll=[]
for i in range(len(t)):
    if i==0:
        ll.append(1)
    else:
        ll.append(t[0]/t[i])
plt.plot(no,ll)
plt.title('Speedup for a fixed array of 2**20');
plt.xlabel('Threads')
plt.ylabel('Speedup')
plt.grid()
```

```python
cost=[]
for i in range(len(t)):
        cost.append(t[i]*no[i])
plt.plot(no,cost)
plt.title('Cost for a fixed array of 2**20');
plt.xlabel('Threads')
plt.ylabel('Cost')
plt.grid()
```

```python
eff=[]
for i in range(len(t)):
        eff.append((t[0]/(t[i]*no[i]))*100)
plt.plot(no,eff)
plt.title('Efficiency for a fixed array of 2**20');
plt.xlabel('Threads')
plt.ylabel('Efficiency (%)')
plt.grid()
```
